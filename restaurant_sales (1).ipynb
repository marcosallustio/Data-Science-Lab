### Merge

```python
R000 = pd.merge(R000, montebello, on='data', how='left')
R001 = pd.merge(R001, piacenza, on='data', how='left')
R002 = pd.merge(R002, piacenza, on='data', how='left')
R003 = pd.merge(R003, piacenza, on='data', how='left')
R004 = pd.merge(R004, stradella, on='data', how='left')
R005 = pd.merge(R005, voghera, on='data', how='left')
```

### Definitive Dataset


```python
r0 = R000
r1 = R001
r2 = R002
r3 = R003
r4 = R004
r5 = R005
```

```python
restaurant = pd.concat([r0,r1,r2,r3,r4,r5])
```

```python
restaurant.reset_index(drop=True, inplace=True)
```

```python
restaurant = restaurant[(restaurant['data']>= "2019-01-01") & (restaurant['data']< "2023-05-01")]
restaurant
```

```python
restaurant['nebbia']=0
restaurant['pioggia']=0
restaurant['temporale']=0
restaurant['neve']=0
restaurant['sereno']=0
restaurant
```

```python
restaurant['FENOMENI'] = restaurant['FENOMENI'].astype('str')
```

```python
restaurant['FENOMENI']
```

```python
restaurant['nebbia'].mask(restaurant['FENOMENI'].str.contains(r'\bnebbia\b'), other=1 , inplace=True)
restaurant['neve'].mask(restaurant['FENOMENI'].str.contains(r'\bneve\b'), other=1 , inplace=True)
restaurant['pioggia'].mask(restaurant['FENOMENI'].str.contains(r'\bpioggia\b'), other=1 , inplace=True)
restaurant['temporale'].mask(restaurant['FENOMENI'].str.contains(r'\btemporale\b'), other=1 , inplace=True)
restaurant['sereno'].mask(restaurant['FENOMENI'].str.contains(r'nan'), other=1 , inplace=True)
```

```python
restaurant[['nebbia', 'neve', 'temporale', 'pioggia', 'sereno']]
```

```python
restaurant['nebbia'].mask(restaurant['HI'].isna(), other= "nan", inplace=True)
restaurant['neve'].mask(restaurant['HI'].isna(), other= "nan", inplace=True)
restaurant['pioggia'].mask(restaurant['HI'].isna(), other= "nan", inplace=True)
restaurant['temporale'].mask(restaurant['HI'].isna(), other= "nan", inplace=True)
restaurant['sereno'].mask(restaurant['HI'].isna(), other= "nan", inplace=True)
restaurant[restaurant['HI'].isna()]
```

```python
restaurant['ristorante'] = restaurant['ristorante'].str.replace(r'00' , '')
restaurant
```
